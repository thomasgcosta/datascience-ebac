1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
````

````python
renda = pd.read_csv("previsao_de_renda.csv")
renda.drop("Unnamed: 0",axis=1,inplace=True)
renda.head()
````

````python
renda.shape
````

````python
ax = sns.countplot(x='data_ref',  hue = 'tipo_renda', data=renda)

tick_labs = renda['data_ref'].map(lambda ts: datetime.strptime(ts,"%Y-%m-%d").strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'educacao', data=renda)

tick_labs = renda['data_ref'].map(lambda ts: datetime.strptime(ts,"%Y-%m-%d").strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref',  hue = 'estado_civil', data=renda)

tick_labs = renda['data_ref'].map(lambda ts: datetime.strptime(ts,"%Y-%m-%d").strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

````python
ax = sns.pointplot(x="data_ref", y="mau", hue = 'tipo_renda',
                   data=renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)

tick_labs = renda['data_ref'].map(lambda ts: datetime.strptime(ts,"%Y-%m-%d").strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(renda['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.50),ncol = 3)
````
